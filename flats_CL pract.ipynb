Выражение для поиска станций метро: (([А-Я]+)|(([А-Я][а-я]+)))\s?((м|М)\.)

```python
tmp = re.sub("(([А-Я]+)|(([А-Я][а-я]+)))\s?((м|М)\.)", "\\5 \\1", flats)
print(tmp)
```

Запишем улицы и переулки в формате "ул./пер. Толмакова".

```python
tmp = re.sub("((\s[А-Я]\.)?\s?[А-Я][а-я]+)(\s(ул\.|пер\.))","\\3\\1",tmp)
tmp = re.sub("(ул\.|пер\.)([А-Я]([а-я]+)?)","\\1 \\2",tmp)
print(tmp)
```

Выведем все адреса домов: ((ул\.|пер\.)\s([А-Я].+)\,?\s(д\.|дом)\s?\d+)

```python
house = re.findall("((ул\.|пер\.)\s([А-Я].+)\,?\s(д\.|дом)\s?\d+)", tmp)
house
```

Выражение для поиска метража квартир: (\d+(((\.|\,)\d)?)\/)?(\d+(((\.|\,)\d)?)\/\d+((\.|\,)\d)?)(\s(кв\.м)|\,\sку)

```python
square = re.findall("(\d+(((\.|\,)\d)?)\/)?(\d+(((\.|\,)\d)?)\/\d+((\.|\,)\d)?)(\s(кв\.м)|\,\sку)", tmp)
square
```
